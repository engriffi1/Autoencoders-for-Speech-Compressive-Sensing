```python
from tensorflow import keras
#from tensorflow.keras.datasets import mnist
```

```python
X_train,X_test=dat[:2500],dat[2500:]
```

```python
X_train1=X_train[:,np.newaxis,:]
X_train1.shape
```

```python
encoder=keras.models.Sequential([
#                                    keras.layers.LSTM(1500,activation=keras.layers.LeakyReLU(alpha=1),input_shape=(X_train1.shape[1], X_train1.shape[2]))
#                                    keras.layers.LSTM(1000,activation=keras.layers.LeakyReLU(alpha=1))
                                 keras.layers.Dense(1500,input_shape=[1,4999],activation=keras.layers.LeakyReLU(alpha=1)),
                                 keras.layers.Dense(1000,activation=keras.layers.LeakyReLU(alpha=1)),
                                 keras.layers.Dense(500,activation=keras.layers.LeakyReLU(alpha=1))
                                 ])
```

```python
decoder=keras.models.Sequential([
#                                     keras.layers.RepeatVector(X_train1.shape[1]),
# #                                     keras.layers.LSTM(1000,activation=keras.layers.LeakyReLU(alpha=1),return_sequences=True),
#                                     keras.layers.LSTM(1500,activation=keras.layers.LeakyReLU(alpha=1),return_sequences=True),
#                                     keras.layers.TimeDistributed(keras.layers.Dense(X_train1.shape[2],activation=keras.layers.LeakyReLU(alpha=1)))
                                 keras.layers.Dense(1000,input_shape=[500],activation=keras.layers.LeakyReLU(alpha=1)),
                                 keras.layers.Dense(1500,activation=keras.layers.LeakyReLU(alpha=1)),
                                 keras.layers.Dense(1*4999,activation=keras.layers.LeakyReLU(alpha=1)),
                                 keras.layers.Reshape([1,4999])
])
```

```python
stacked_ae=keras.models.Sequential([encoder,decoder])
```

```python
stacked_ae.compile(loss="mae",optimizer=keras.optimizers.SGD(lr=1.5))
stacked_ae.summary()
```
